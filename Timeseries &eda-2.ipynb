```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load aligned data
df = pd.read_csv("data/Aligned_Data.csv")

# Basic statistics
print(df.describe())

# Correlation
print(df[['GDP_per_Capita_USD', 'Internet_Usage_%']].corr())

# Scatter plot
sns.scatterplot(data=df,
                x='Internet_Usage_%',
                y='GDP_per_Capita_USD')

plt.title("Internet Usage vs GDP per Capita")
plt.show()

# Time trend (first 5 countries)
for country in df['Country Name'].unique()[:5]:
    country_data = df[df['Country Name'] == country]
    plt.plot(country_data['Year'],
             country_data['GDP_per_Capita_USD'],
             label=country)

plt.legend()
plt.title("GDP Trend Over Time")
plt.show()

print("EDA completed.")
```
